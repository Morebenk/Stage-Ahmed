# 05 - Effective Notch Stress Method

Demonstrates the Effective Notch Stress Method per IIW.

- Reference radius: r_ref = 1 mm (standard), 0.05 mm (thin sheets < 5mm)
- Single FAT class: **FAT 225** (steel), **FAT 71** (aluminum)
- Applicable to all weld geometries

```python
from weldfatigue.fatigue.notch_stress import EffectiveNotchStressAssessment
from weldfatigue.fatigue.sn_curve import SNCurve
import matplotlib.pyplot as plt
```

## Standard Plate (t ≥ 5mm)

```python
notch = EffectiveNotchStressAssessment(material="steel", plate_thickness=10.0)
print(f"FAT class: {notch.fat_class}")
print(f"Reference radius: {notch.r_ref} mm")

result = notch.evaluate(stress_range=200.0, num_cycles=1_000_000)
print(f"\nΔσ_notch = 200 MPa, N = 10^6")
print(f"Allowable cycles: {result.allowable_cycles:.2e}")
print(f"Safety factor: {result.safety_factor:.3f}")
print(f"Status: {result.status}")
```

## Thin Sheet (t < 5mm)

```python
notch_thin = EffectiveNotchStressAssessment(material="steel", plate_thickness=3.0)
print(f"FAT class: {notch_thin.fat_class}")
print(f"Reference radius: {notch_thin.r_ref} mm")

result_thin = notch_thin.evaluate(stress_range=200.0, num_cycles=1_000_000)
print(f"Allowable cycles: {result_thin.allowable_cycles:.2e}")
print(f"Status: {result_thin.status}")
```

## Aluminum Notch Assessment

```python
notch_al = EffectiveNotchStressAssessment(material="aluminum", plate_thickness=10.0)
print(f"FAT class: {notch_al.fat_class} (aluminum)")

result_al = notch_al.evaluate(stress_range=60.0, num_cycles=1_000_000)
print(f"Δσ_notch = 60 MPa → Allowable = {result_al.allowable_cycles:.2e}")
print(f"Status: {result_al.status}")
```

## Compare: Notch FAT 225 vs Nominal FAT 80

```python
sn_notch = SNCurve(fat_class=225, material_type="steel")
sn_nominal = SNCurve(fat_class=80, material_type="steel")

fig, ax = plt.subplots(figsize=(10, 6))
n1, s1 = sn_notch.get_curve_points()
n2, s2 = sn_nominal.get_curve_points()
ax.loglog(n1, s1, 'b-', linewidth=2, label='FAT 225 (Notch)')
ax.loglog(n2, s2, 'r--', linewidth=2, label='FAT 80 (Nominal)')
ax.set_xlabel("N cycles")
ax.set_ylabel("Δσ [MPa]")
ax.set_title("Notch (FAT 225) vs Nominal (FAT 80) S-N Curves")
ax.legend()
ax.grid(True, which='both', alpha=0.3)
plt.tight_layout()
plt.show()
```
