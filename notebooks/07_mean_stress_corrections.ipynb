# 07 - Mean Stress Corrections

Compare Goodman, Gerber, Soderberg corrections and the IIW f(R) enhancement factor.

| Method | Formula |
|--------|--------|
| Goodman | $S_{a,eq} = \frac{S_a}{1 - S_m / S_u}$ |
| Gerber | $S_{a,eq} = \frac{S_a}{1 - (S_m / S_u)^2}$ |
| Soderberg | $S_{a,eq} = \frac{S_a}{1 - S_m / S_y}$ |
| IIW f(R) | $f(R) = -0.4R + 1.2$ for $-1 \le R \le 0.5$ |

```python
from weldfatigue.fatigue.mean_stress import MeanStressCorrection
from weldfatigue.reporting.plots import FatiguePlots
import numpy as np
import matplotlib.pyplot as plt
```

## Compare Corrections at Various Mean Stresses

```python
Sa = 80.0  # stress amplitude
Su = 600.0  # ultimate strength
Sy = 350.0  # yield strength

mean_stresses = np.arange(0, 300, 25)
print(f"{'Sm [MPa]':>10}  {'Goodman':>10}  {'Gerber':>10}  {'Soderberg':>10}")
print("-" * 45)
for sm in mean_stresses:
    g = MeanStressCorrection.goodman(Sa, sm, Su)
    ge = MeanStressCorrection.gerber(Sa, sm, Su)
    s = MeanStressCorrection.soderberg(Sa, sm, Sy)
    print(f"{sm:>10.0f}  {g:>10.2f}  {ge:>10.2f}  {s:>10.2f}")
```

## Haigh Diagram

```python
Se = 100.0  # endurance limit

operating_points = [(50, 80), (100, 70), (200, 50), (150, 90)]
fig = FatiguePlots.haigh_diagram_plotly(
    Se=Se, Su=Su, Sy=Sy,
    operating_points=operating_points,
)
fig.show()
```

## IIW f(R) Enhancement Factor

```python
R_values = np.linspace(-1, 0.5, 50)

fig, ax = plt.subplots(figsize=(8, 5))
f_values_relieved = [MeanStressCorrection.iiw_enhancement_factor(r, 'stress_relieved') for r in R_values]
f_values_welded = [MeanStressCorrection.iiw_enhancement_factor(r, 'as_welded') for r in R_values]

ax.plot(R_values, f_values_relieved, 'b-', linewidth=2, label='Stress-relieved')
ax.plot(R_values, f_values_welded, 'r--', linewidth=2, label='As-welded')
ax.set_xlabel('Stress ratio R', fontsize=12)
ax.set_ylabel('f(R)', fontsize=12)
ax.set_title('IIW Enhancement Factor f(R)', fontsize=14)
ax.legend(fontsize=11)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## Stress Ratio Computation

```python
print("Stress ratio R = σ_min / σ_max")
print(f"R(-1) fully reversed:  R = {MeanStressCorrection.stress_ratio(-100, 100):.2f}")
print(f"R(0)  pulsating:       R = {MeanStressCorrection.stress_ratio(0, 100):.2f}")
print(f"R(0.5) positive mean:  R = {MeanStressCorrection.stress_ratio(50, 100):.2f}")
```
